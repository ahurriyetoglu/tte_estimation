```python
def tokenize(tw):
    newtw = ''
    lentw = len(tw)
    for i, c in enumerate(tw):
        if (c in "'`´’‘") and ((i+1 != lentw) and (i!=0)) and ((tw[i-1].isalpha()) or tw[i-1] in "0123456789") and (tw[i+1].isalpha()):
            newtw += ' '+c+' '
        elif (c in "'`´’()+*-") and ((lentw>i+1) and (i>0)) and (tw[i-1] in ' 0123456789') and (tw[i+1].isalpha()):
            newtw += c+' '
        elif (c in '();>') and ((lentw>i+1) and (i!=0)) and ((tw[i-1].isalpha()) or (tw[i-1] in '0123456789')) and ((tw[i+1] == ' ') or (i == lentw-1)):
            newtw += ' '+c
        elif (c in "'`´’‘()+*->") and (i==0) and (lentw > 1) and ((tw[1].isalpha()) or tw[1] in "0123456789"):
            newtw += c+' '
        elif (c in "'`´’‘()+*->") and (i+1 == lentw) and (lentw > 1) and ((tw[i-1].isalpha()) or tw[i-1] in "0123456789"):
            newtw += ' '+c
        elif (c in ",") and ((i != 0) and (i+1 != lentw)) and tw[i-1].isdigit() and tw[i+1].isdigit(): # for 3,5 7,5
            newtw += c
        elif (c in ","):
            newtw += " "+c+" "
        elif (c in "â"):
            newtw += "a"
        elif (c in "ê"):
            newtw += "e"
        elif (c in "î"):
            newtw += "i"
        elif (c in "ú"):
            newtw += "ü"
        else:
            newtw += c
        #print(c in "'`´’()+*-", lentw>i+1, i>0, tw[i-1] == ' 0123456789', tw[i+1].isalpha())
            
    return newtw
```

```python
dtc = datetime.datetime.combine # for time difference
dty = datetime.date.today() # for day difference

def getlinelist(filename, striplines=True):
    with open(filename) as f:
        linelist = f.read().splitlines()
    
    if striplines:
        linelist = [l.strip() for l in linelist]
        
    return [l for l in linelist if len(l)>0] # Exclude blank lines!
    

def writelinelist(filename, mylist):
    with open(filename, 'w') as fw:
        for l in mylist[:-1]: # Write lines with new line, except the last line.
            fw.write(str(l)+'\n')
        fw.write(str(mylist[-1])) # Do not put new line for the last line!
        
def isMweAlpha(mwe): # True for mwe's which does not have any numbers in them
    for m in mwe.split():
        if not m.isalpha():
            return False
    return True

def is_outlier(points, thresh=3):
    """
    Returns a boolean array with True if points are outliers and False otherwise.

    Parameters:
    -----------
        points : An numobservations by numdimensions array of observations
        thresh : The modified z-score to use as a threshold. Observations with
            a modified z-score (based on the median absolute deviation) greater
            than this value will be classified as outliers.

    Returns:
    --------
        mask : A numobservations-length boolean array.

    References:
    ----------
        Boris Iglewicz and David Hoaglin (1993), "Volume 16: How to Detect and Handle Outliers", The ASQC Basic 
        References in Quality Control: Statistical Techniques, Edward F. Mykytka, Ph.D., Editor. 
    """
    if len(points.shape) == 1:
        points = points[:,None]
    median = np.median(points, axis=0)
    diff = np.sum((points - median)**2, axis=-1)
    diff = np.sqrt(diff)
    med_abs_deviation = np.median(diff)

    modified_z_score = 0.6745 * diff / med_abs_deviation

    return modified_z_score > thresh

def scatter_dict(view, name, d):
    """partition a dictionary across the engines of a view"""
    # Ex use: scatter_dict(dview, 'test', test_dict)
    ntargets = len(view)
    keys = list(d.keys()) # d.keys() in Python 2
    for i, target in enumerate(view.targets):
        subd = {}
        for key in keys[i::ntargets]:
            subd[key] = d[key]
        view.client[target][name] = subd

def gather_dict(view, name):
    """gather dictionaries from a DirectView"""
    #Ex. use: gather_dict(dv, 'test')
    merged = {}
    for d in view.pull(name):
        merged.update(d)
    return merged

#Eliminate RT's
def getNonRT(etable):
    return etable[~etable.ttext.str.contains("rt @", case=False)].copy() # ~ not
# Get just RT's
def getRT(etable):
    return etable[table.ttext.str.contains("rt @", case=False)].copy() # ~ not

# calculate time to event
def time_to_event(groupForEvent, def_etimedict = etimedict):
    t_toevent = def_etimedict[groupForEvent.name] - groupForEvent.dt_etime
    groupForEvent['time_to_event'] = t_toevent
    
    return groupForEvent

def getBasicWordSet(mylist):
    # input: list of strings
    # output: set of words in this list
    myset = set()
    for tw in mylist:
        if isinstance(tw, str):
            tw = tw.split()
        for w in tw:
            myset.add(w)
    return myset

def listoverlap(t1, t2):
    for t in t1:
        if t in t2:
            return True
    return False

def getWordSet(elimPlace=True,elimStop=True,elimForeign=True, elimTmx=False):
    # Place names
    geonamesNL = pd.read_table('../timexNelleke/geonamesNL.txt', sep='\t', usecols=[1], names=['pName'])
    geonamesNL.pName = geonamesNL.pName.str.lower()
    
    otbasisongekeurd = getlinelist('../timexNelleke/OpenTaal-210G-basis-ongekeurd.txt')
    otbasisongekeurd = [otbg.lower() for otbg in otbasisongekeurd] # place names to exclude
    
    manualPlaceNames = getlinelist("../timexNelleke/20140318-placeNames.txt") # place names to exclude
    
    geonamesNL_list = list(geonamesNL.pName) + manualPlaceNames + otbasisongekeurd
    
    #stop and foreign words
    foreignWords = getlinelist("../timexNelleke/20140318-foreignWords.txt")
    stopwords = getlinelist('../timexNelleke/20140318-stopwords.txt') + foreignWords
    
    # Read the Lexicon
    otbasisgekeurd = getlinelist('../timexNelleke/OpenTaal-210G-basis-gekeurd.txt') 
    otflexiformen = getlinelist('../timexNelleke/OpenTaal-210G-flexievormen.txt')
    otset =  set([otbg.lower() for otbg in otbasisgekeurd+otflexiformen])
    
    occTmx = getlinelist("../timexNelleke/20140318-occTmxAllEvAndTime.txt")
    
    # Eliminate various word subsets
    if elimPlace:
        otset = otset - set(geonamesNL_list) # eliminate place names from geonames NL
        
    if elimStop:
        otset = otset - set(stopwords) # use content words only, eliminate stop words
    
    if elimForeign:
        otset = otset - set(foreignWords)
        
    if elimTmx:
        otset = otset - set(occTmx) # eliminate occurred time expressions
    
    return otset

def getOccWords(etable, otset, thres=1):
    # Find word frequencies if they occur in opentaal list
    # FAST:use dict, counter, set intersection.
    word_set = set()
    word_counter = Counter()
    for w_list in etable.ttextlist:
        for w in w_list:
            word_set.add(w)
    occ_w_list = list(otset & word_set)
    
    word_counter = Counter()
    for w in occ_w_list:
        word_counter[w] = 0 # needed to not to include words which are not in occ_w_list
    
    for word_list in etable.ttextlist:
        for w in word_list:
            if w in word_counter: #is alpha eliminates multi word expressions
                word_counter[w] += 1
                
    return [k for k, v in word_counter.most_common() if v>thres] # eliminate 1 time occurring words

def extractFeats(experiment, use_wlist):
    use_wlist = list(set(use_wlist))
    use_wlist = sorted(use_wlist, key=len, reverse=True)
    use_wlist = [tx.replace('.','\.') for tx in use_wlist] # not to match anything with . (dot)
    wlist_ptrn = re.compile('\\b'+'\\b|\\b'.join(use_wlist)+'\\b')
    
    # Do parallel extraction
    dview.scatter("experiment", experiment)
    dview["wlist_ptrn"] = wlist_ptrn
    dview.execute("experiment['allFeats'] = experiment.ttext.str.findall(wlist_ptrn)")
    return pd.concat(dview.gather("experiment"))

def creaWserie(wList, lComb):# = lenComb
    ''' Provide all combinations in the same order up to length of lcomb'''
    allFeatures = chain.from_iterable([combinations(wList,i) for i in range(1, lComb+1)]) # not all lengths:range(1,len(wList)+1)
    allFeatures = ["_".join(x) for x in allFeatures]
    return allFeatures

def getTrainFeatures(trainData, thresQuantile=0.90, thresOcc=1, excAmbTemp=False, rulelist=[]):
    
    excWFile="/home/hurrial/Dropbox/timexNelleke/20140601-excludeTmx.txt"
    if getpass.getuser() == 'alihurriyetoglu':
        excWFile="/Users/alihurriyetoglu/Dropbox/timexNelleke/20140601-excludeTmx.txt"        
    
    #print(excWFile)
    exclude_words_set=set(getlinelist(excWFile))
    #exclude_words_set=set(["al","zo","dan","nog"])
    featCnt = Counter()
    for tc in trainData["featComb"]:
        for t in tc:
            featCnt[t] += 1
    print("len features 1:", len(featCnt))
    
    #print("Rule List:",rulelist)
    #if a feature occur more than 1, include it
    featDict = {}
    for k, v in featCnt.most_common():
        if v>thresOcc: # featSelect == 0 cause the quantile based elimination to behave strange due to many 0's
            if (excAmbTemp and len(set(k.split('_')) - exclude_words_set)==0) or (k in rulelist):
                #exclude temporal features which consist only ambiguous items!
                # do not do any calculation for the features which has rules.
                continue
            featDict[k] = []
    print("len features 2:", len(featDict))
    featDictStd, featDictMedAD, featDictMed, featDictMea, featDictTSerie = {}, {}, {}, {}, {}
    
    # gather tte for the features
    for wCmb, tHF in zip(trainData.featComb, trainData.time_to_eventhF):
        for wC in wCmb:
            if wC in featDict:
                featDict[wC].append(tHF)
    
    for feat, wtserie in featDict.items():
        wtserie = np.array(wtserie)
        featDictStd[feat] = np.std(wtserie)
        featDictMedAD[feat] = np.median(np.fabs(wtserie - np.median(wtserie)))
    
    featSelect = pd.Series(featDictStd)
    featSelectMedAD = pd.Series(featDictMedAD)
    
    featSelect = featSelect[featSelect > 0] # should be done before quantile based elimination!!!
    featSelectMedAD = featSelectMedAD[featSelectMedAD > 0]
    
    #workon5
    if thresQuantile < 1:
        train_thresQuantile = featSelect.quantile(thresQuantile)
        
        if False:
            print("cut-off std is:",train_thresQuantile)
            elimFeats = featSelect[featSelect>train_thresQuantile]
            print("\neliminated:",len(elimFeats))
            for k, v in featCnt.most_common():
                if k in elimFeats.index:
                    print(k, elimFeats.ix[k], v)
                    
            usedFeats=featSelect[featSelect<train_thresQuantile]
            print("\nUsed Features:",len(usedFeats))
            for k, v in featCnt.most_common():
                if k in usedFeats.index:
                    print(k, usedFeats.ix[k], v)
            
        featSelect = featSelect[featSelect<train_thresQuantile]
        
        train_thresQuantile = featSelectMedAD.quantile(thresQuantile)
        featSelectMedAD = featSelectMedAD[featSelectMedAD<train_thresQuantile]
        
    print("len features 3,std, medAD:", len(featSelect), len(featSelectMedAD))
    
    featDict2 = {}
    for fs in featSelect.index:
        featDict2[fs] = featDict[fs]
    #featDict = {fs:featDict[fs] for fs in featSelect.index}
    
    featDict3 = {}
    for fs in featSelectMedAD.index:
        featDict3[fs] = featDict[fs]
                
#     #featuresDF = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in featDict.items()]))
    for feat, wtserie in featDict2.items():
    #for feat, wtserie in featDict3.items():
        featDictTSerie[feat] = np.array(wtserie) 
        featDictMed[feat] = np.median(featDictTSerie[feat])
        featDictMea[feat] = np.mean(featDictTSerie[feat])
        featDictTSerie[feat] = np.round(featDictTSerie[feat],2)
        
        #featDictMea[feat] = np.sqrt(np.dot(np_wtserie,np_wtserie)/len(np_wtserie))
    
    return pd.DataFrame({"meanTTE":featDictMea, "medianTTE":featDictMed, "TSerie":featDictTSerie})
```

```python
# extract used features and check how much do they contribute to the performance!
def getFeatPerfDF(resultDF, colname):
    resultDF["error"] = resultDF.time_to_eventhF - resultDF[colname]
    usedFeatDict={}
    for usFts, estErr in zip(resultDF.usedFeats,resultDF.error):
        for uf in usFts:
            if uf not in usedFeatDict:
                usedFeatDict[uf]=[]
                #print(uf)
            usedFeatDict[uf].append(estErr)
    #tstResultDFW_usedFeats. [["featComb","usedFeats","error"]]

    usedFeatPerfDF = pd.DataFrame()#{"mean_error":300,"frequency":1}, index=['startxxx']
    for usedFeat, featErrs in usedFeatDict.items():
        usedFeatPerfDF=usedFeatPerfDF.append(pd.DataFrame({"mean_error":numpy.mean(numpy.abs(featErrs)), "frequency":len(featErrs)}, index=[usedFeat]))

    try:
        usedFeatPerfDF["mean_error_int"]=usedFeatPerfDF.mean_error.astype(int)
        return usedFeatPerfDF.sort(["mean_error_int", "frequency"], ascending=[1,0])
    except:
        return usedFeatPerfDF
```

```python
def getHyperHist(orgDF, estCol, uptoWLen=51, step=3):
    resultsDFTRDevHist = orgDF[orgDF[estCol].notnull()]
    #print(len(resultsDFTRDevHist))
    hstWlnErrDF = pd.DataFrame({"medMAE":getMAE(resultsDFTRDevHist.time_to_eventhF, resultsDFTRDevHist[estCol]), 
                                                       "medRMSE":getRMSE(resultsDFTRDevHist.time_to_eventhF, resultsDFTRDevHist[estCol]),
                                                       "meaMAE":getMAE(resultsDFTRDevHist.time_to_eventhF, resultsDFTRDevHist[estCol]),
                                                       "meaRMSE":getRMSE(resultsDFTRDevHist.time_to_eventhF, resultsDFTRDevHist[estCol])}, index=[1])
    #print(resultsDFTRDevHist.event.value_counts())
    print('wlen', end=':')
    for wlen in range(2,uptoWLen,step):
        # groupby("event",group_keys=False)
        resultsDFTRDevHist = resultsDFTRDevHist.groupby("event",group_keys=False).apply(calcExpandHistNorm, estCol, wlen)
        hstWlnErrDF = hstWlnErrDF.append(pd.DataFrame({"medMAE":getMAE(resultsDFTRDevHist.time_to_eventhF, resultsDFTRDevHist["histMed_"+estCol]), 
                                                       "medRMSE":getRMSE(resultsDFTRDevHist.time_to_eventhF, resultsDFTRDevHist["histMed_"+estCol]),
                                                       "meaMAE":getMAE(resultsDFTRDevHist.time_to_eventhF, resultsDFTRDevHist["histMea_"+estCol]),
                                                       "meaRMSE":getRMSE(resultsDFTRDevHist.time_to_eventhF, resultsDFTRDevHist["histMea_"+estCol])}, index=[wlen]))
        print(wlen, end=',')
    print()
    return hstWlnErrDF
```

```python
def p_corr(df1, df2):
    """
    Computes Pearson correlation and its significance (using a t
    distribution) on a pandas.DataFrame.
 
    Ignores null values when computing significance. Based on
    http://en.wikipedia.org/wiki/Pearson_product-moment_correlation_coefficient#Testing_using_Student.27s_t-distribution
 
    Args:
        df1 (pandas.DataFrame): one dataset
        df2 (pandas.DataFrame): another dataset
 
    Returns:
        corr (float): correlation between the two datasets
        t (float): an associated t-value
        p (float): one-tailed p-value that the two datasets differ
    """
    corr = df1.corr(df2)
    N = np.sum(df1.notnull())
    t = corr*np.sqrt((N-2)/(1-corr**2))
    p = 1-scipy.stats.t.cdf(abs(t),N-2)  # one-tailed
    return corr, t, p
```
